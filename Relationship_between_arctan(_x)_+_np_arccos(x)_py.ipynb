```python
import numpy as np
import matplotlib.pyplot as plt

# x 값의 범위와 간격 설정
x = np.linspace(-10, 10, 100)

# y 값 계산: arcsin(x) + arccos(x)
y = np.arctan(-x) + np.arccos(x)

# 선 그래프 그리기
plt.plot(x, y, label="arctan(-x) + np.arccos(x)")

# pi/2 직선 추가 (예상 결과)
plt.axhline(y=np.pi/2, color='r', linestyle='--', label="π/2")

# x축과 y축의 레이블 설정
plt.xlabel('x')
plt.ylabel('y')

# 그래프 제목 및 범례 설정
plt.title('Relationship between arctan(-x) + np.arccos(x)')
plt.legend()

# 그래프 보여주기
plt.show()
```
